```python
import torch
import torch as np
```

```python
# Initialize the tensor
tensor_a = torch.tensor([1, 2, 3, 4])
tensor_b = torch.tensor([5, 6, 7, 8])
```

```python
# Addition
tensor_add = tensor_a + tensor_b
tensor_add
```

```python
# Addition Long hand
tensor_add1 = torch.add(tensor_a, tensor_b)
tensor_add1
```

```python
# subtract
tensor_subtract = tensor_a - tensor_b
tensor_subtract
```

```python
# Sub function
torch.sub(tensor_a, tensor_b)
```

```python
# Multiplication
tensor_mult = tensor_a * tensor_b
tensor_mult
```

```python
# Multiplicaton Long hand
torch.multiply(tensor_a, tensor_b)
```

```python
# Division
tensor_b / tensor_a
```

```python
# Division long hand
torch.div(tensor_b, tensor_a)
```

```python
# Modulus
tensor_b % tensor_a
```

```python
# Modulus Long hand
torch.remainder(tensor_b, tensor_a)
```

```python
# Exponents
torch.pow(tensor_b, tensor_a)
```

```python
# Another way to write long hand
tensor_a.add(tensor_b)
```

```python
# Reassignment
tensor_a = tensor_a + tensor_b
tensor_a
```
