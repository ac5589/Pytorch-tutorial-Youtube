```python
import torch
```

```python
# create a tensor
my_torch = torch.arange(10)
my_torch
```

```python
# Reshape and view the tensor
my_torch = my_torch.reshape(2, 5)
my_torch
```

```python
# reshpae if we dont know the number of items
my_torch2 = torch.arange(10)
my_torch2 = my_torch2.reshape(2, -1)
my_torch2
```

```python
my_torch3 = torch.arange(10)
my_torch3 = my_torch3.view(2, 5)
my_torch3
```

```python
# reshape and view does update the final tensor
my_torch5 = torch.arange(10)
my_torch6 = my_torch5.reshape(2, 5)
my_torch6
my_torch5[1] = 414
my_torch6
```

```python
# Grab a specifc item
my_torch7 = torch.arange(10)
my_torch7[7]
```

```python
# Grab a column
my_torch8 = torch.arange(10)
my_torch8 = my_torch8.reshape(5, 2)
my_torch8[:, 1]
```

```python
# Return a column
my_torch8 = torch.arange(10)
my_torch8 = my_torch8.reshape(5, 2)
my_torch8[:, 1:]
```
