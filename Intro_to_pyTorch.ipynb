```python
# Building a model from scratch

# Model: simple linear regression
# y = xW + b
def model(X, W, b):
  return X @ W + b

# first, create a fake data

# Number of data points in the input data
N = 10

# Each data point has 1 input feature and 1 output value
D_in = 1;
D_out = 1;

# Create our input data
X = torch.randn(N, D_in)

true_W = torch.tensor([[2.0]])
true_b = torch.tensor(1.0)
y_true = model(X, true_W, true_b) + torch.randn(N, D_out) * 0.1

# Initialize our parameters with random values
W = torch.randn(D_in, D_out, requires_grad=True)
b = torch.randn(1, requires_grad=True)

# Print the weights
print(f'Initial Weight W = \n{W}\n')
print(f'Initial Bias b = \n{b}\n')

print(model(X, W, b))

# Root mean squared error
loss = ((model(X, W, b) - y_true)**2).mean()
print(loss)

# Learning the weight and bias to reduce the error
loss.backward()

# Gradients are stored in the .grad attribute
print(f'Gradient for W: \n {W.grad}\n')
print(f'Gradient for b: \n {b.grad}\n')
```

```python
# Train the model via manually update the layers

# Learning rate and epoch
lr, epochs = 0.01, 1000

# Initialize the parameters
W, b = torch.randn(1, 1, requires_grad=True), torch.randn(1, requires_grad=True)

# Iterate the weights and bias via a for loop

for epoch in range(epochs):

  # Get the prediction
  # forward Pass
  y_hat = X @ W + b

  # Get the Loss function
  loss = torch.mean(((y_hat - y_true)**2))

  # Backward Pass
  loss.backward()

  # Update the parameters
  with torch.no_grad():
    W -= lr*W.grad
    b -= lr*b.grad

  # Initialize the gradient to zero
  W.grad.zero_(); b.grad.zero_()

  if epoch%10 == 0:
    print(f'Epoch = {epoch}, W = {W.item():.3f}, b = {b.item():.3f}, Loss = {loss.item():.3f}')

print(f'True parameters, W = {true_W.item():.3f}, b = {true_b.item():.3f}')
print(f'Final learned parameters, W = {W.item()}, b = {b.item()}')

```

```python
# Train the model via torch.nn model

# Input as one feature
# Dimension of the input = 1
D_in = 1

# Output as one value
# dimension of the output = 1
D_out = 1

# Create the Linear layer
Linear_layer = torch.nn.Linear(in_features=D_in, out_features=D_out)

print(f'Layers Weight (W) {Linear_layer.weight}')
print(f'Layers Bias (b) {Linear_layer.bias}')

# Forward Pass
y_hat_nn = Linear_layer(X)

print(f'Ouput of nn.Linear {y_hat_nn}')
```
