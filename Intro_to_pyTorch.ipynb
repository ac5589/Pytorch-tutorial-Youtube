```python
# Every tensor has three atributes
tensor = torch.randn(2, 3)
print(f'Shape: {tensor.shape}')
print(f'Datatype: {tensor.dtype}')
print(f'Device: {tensor.device}')
```

```python
# Automatic differntiation aka auto grad

x_data = torch.tensor([[1., 2,], [3., 4.]])

w = torch.tensor([[1.0, 2.0], [3.0, 4.0]], requires_grad=True)

print(f'x_data = {x_data.requires_grad}')
print(f'w = {w.requires_grad}')
```

```python
# Goal: compute Z = x*y, where y = a + b

a = torch.tensor(2.0, requires_grad=True)
b = torch.tensor(3.0, requires_grad=True)
x = torch.tensor(4.0, requires_grad=True)

y = a + b

z = x * y

print(f'grad_fn for z = {z.grad_fn}')
print(f'grad_fn for y = {y.grad_fn}')
print(f'grad_fn for a = {a.grad_fn}')
```

```python
# Building a model from scratch

# Model: simple linear regression
# y = xW + b

# first, create a fake data

# Number of data points in the input data
N = 10

# Each data point has 1 input feature and 1 output value
D_in = 1;
D_out = 1;

# Create our input data
X = torch.randn(N, D_in)

true_W = torch.tensor([[2.0]])
true_b = torch.tensor(1.0)
y_true = X @ true_W + true_b + torch.randn(N, D_out) * 0.1

# Initialize our parameters with random values
W = torch.randn(D_in, D_out, requires_grad=True)
b = torch.randn(1, requires_grad=True)

print(f'Initial Weight W = \n{W}\n')
print(f'Initial Bias b = \n{b}\n')

def model(X, W, b):
  return X @ W + b

print(model(X, W, b))

loss = ((model(X, W, b) - y_true) **2).mean()

print(loss)


```
